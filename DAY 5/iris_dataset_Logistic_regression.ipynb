```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
model = LogisticRegression()
flower = load_iris()
```

```python
dir(flower)
```

```python
print(flower.feature_names)
```

```python
print(flower.target_names)
```

```python
plt.gray()
for i in range(5):
    plt.matshow(flower.data[i].reshape(2,2))
    plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(flower.data, flower.target, test_size = 0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(flower.data[0:10])
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
