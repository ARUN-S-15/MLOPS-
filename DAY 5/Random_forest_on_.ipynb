```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline
flower = load_iris()
```

```python
dir(flower)
```

```python
plt.gray()
for i in range(5):
  plt.matshow(flower.data[i].reshape(2, 2))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(flower.data)
df.head()
```

```python
df['target'] = flower.target
```

```python
df[0:12]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(flower.data, flower.target, test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 10)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
m = confusion_matrix(y_test, y_predict)
m
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(m, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.svm import SVC
```

REGULARIZATION


```python
model_r = SVC(C = 10)
model_r.fit(x_train, y_train)
model_r.score(x_test, y_test)
```

```python
model_r = SVC(C = 20)
model_r.fit(x_train, y_train)
model_r.score(x_test, y_test)
```

GAMMA

```python
model_g = SVC(gamma = 10)
model_g.fit(x_train, y_train)
model_g.score(x_test, y_test)
```

```python
model_g = SVC(gamma = 20)
model_g.fit(x_train, y_train)
model_g.score(x_test, y_test)
```
