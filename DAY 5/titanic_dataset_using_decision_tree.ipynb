```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
input = df.drop('Survived', axis = 'columns')
```

```python
target = df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
l_name = LabelEncoder()
l_sex = LabelEncoder()
l_ticket = LabelEncoder()
l_cabin = LabelEncoder()
l_embarked = LabelEncoder()
```

```python
input['name_n'] = l_name.fit_transform(input['Name'])
input['sex_n'] = l_sex.fit_transform(input['Sex'])
input['ticket_n'] = l_ticket.fit_transform(input['Ticket'])
input['cabin_n'] = l_cabin.fit_transform(input['Cabin'])
input['embarked_n'] = l_embarked.fit_transform(input['Embarked'])
```

```python
input
```

```python
input_n = input.drop(['Name', 'Sex', 'Cabin', 'Embarked', 'Ticket'], axis = 'columns')
```

```python
input_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(input_n, target)
```

```python
model.score(input_n, target)
```
